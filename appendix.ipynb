```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_lambda(c,n):
    return 0.5*c*n/(c*n+1)
```

```python
def calc_eta(c,n):
    return 0.5**(1/(c*n))
```

```python
c_values = np.arange(0.01, 0.2, 0.01)
n=4
```

```python
plt.plot(c_values, [calc_lambda(c,n) for c in c_values])
plt.plot(c_values, [calc_eta(c,n) for c in c_values])
plt.xlabel('c')
plt.legend(['lambda', 'eta'])
```
